```python
import time 
import pandas as pd 
import requests
from collections import defaultdict
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait 
```

```python
driver=webdriver.Chrome() 
```

```python
website='https://www.amazon.com/'
driver.get(website)
```

```python
computer=driver.find_element(By.XPATH,'//*[@id="CardInstancePo0vaUfBCHCSXnBR-wD8Cg"]/div[3]/a')
computer.click()
time.sleep(0.5)
```

```python


# Initialize the list to store the data
data = []
data.append(['name', 'price', 'image'])  # Adding headers for name, price, and image

i = 1
while True:
    try:
        # Find the product container using XPath for name, price, and image
        product = driver.find_element(By.XPATH, f'//*[@id="search"]/div[1]/div[1]/div/span[1]/div[1]/div[{i}]')

        # Extract the text from the product container (name and price)
        product_text = product.text.split('\n')

        # Extract the name (usually the first element in the split text)
        product_name = product_text[0] if len(product_text) > 0 else 'Name not found'

        # If the first line is 'Best Seller', get the second line as the name
        if product_name == 'Best Seller' and len(product_text) > 1:
            product_name = product_text[1]  # Skip 'Best Seller' and take the actual product name

        # Check if the price contains a range (e.g., "$5 - $8")
        product_price = 'Price not found'
        for line in product_text:
            if '$' in line:  # Check for price-related text
                if '-' in line:  # If price contains a range
                    product_price = line.strip()
                elif '$' in line:  # For fixed price
                    product_price = line.strip()

        # Extract the product image URL
        image_url = ''
        try:
            image_element = product.find_element(By.XPATH, './/span/a/div/img')
            image_url = image_element.get_attribute('src')
        except:
            image_url = 'Image not found'

        # Append the name, price, and image URL to the data list
        data.append([product_name, product_price, image_url])

        # Print for debugging
        print(f"Name: {product_name}, Price: {product_price}, Image URL: {image_url}")

        # Increment the counter to move to the next product
        i += 1

        # Small delay to avoid overwhelming the server
        time.sleep(0.3)

    except Exception as e:
        # Stop if no more products are found or if there's an error
        print(f"Error or No more products found: {e}")
        break

# Save to CSV after scraping
df = pd.DataFrame(data)  # Create a DataFrame
df.to_csv('amazon_products_with_images.csv', index=False)  # Save to CSV without the index
```

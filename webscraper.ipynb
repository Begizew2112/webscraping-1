```python
import time 
import pandas as pd 
import requests
from collections import defaultdict
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
driver=webdriver.Chrome()
```

```python
website='https://webscraper.io/test-sites/e-commerce/more'
driver.get(website)
```

```python
try:
    accept_and_continuew=driver.find_element(By.XPATH,'//*[@id="cookieBanner"]/div[2]/a')
    accept_and_continuew.click()
    time.sleep(0.1)
except:
    pass


```

```python
computer=driver.find_element(By.XPATH,'//*[@id="side-menu"]/li[2]')
computer.click()
time.sleep(0.5)
```

```python
laptop=driver.find_element(By.CSS_SELECTOR,'#side-menu > li.nav-item.active > ul > li:nth-child(1) > a')
laptop.click()
```

```python

while True:
    try:  
        click_more=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div/div[2]/a')
        click_more.click()
        time.sleep(1)
    except:
        break
```

```python
data=[]
data.append(['price','type of laptop',' description about laptop','reviews'])
i=1
while True:
    try:
        list=driver.find_element(By.XPATH,f'/html/body/div[1]/div[3]/div/div[2]/div/div[{i}]')
        data.append(list.text.split('\n'))
        i += 1
        print(list.text)
        time.sleep(0.1)
    except:
        break
    

```

```python
print(len(data)) 
```

```python
 
# Create a DataFrame
df=pd.DataFrame(data[1:],columns=data[0]) 
# Printing the DataFrame
print(df)
# Exporting to an Excel file 
df.to_excel('output.xlsx', index=False) 
```

```python
df = pd.DataFrame(data)

# Specify the output CSV file path
output_csv_file = "output.csv"

# Write the DataFrame to a CSV file
df.to_csv(output_csv_file, index=False, header=False)

print(f"Successfully wrote {len(df)} rows to {output_csv_file}")
```
